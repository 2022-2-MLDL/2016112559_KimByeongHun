```python
import pandas as pd
```

```python
tumordf = pd.read_csv('BC-TCGA-Tumor.txt', sep='\t' )
tumordf.head()
```

```python
normaldf = pd.read_csv('BC-TCGA-Normal.txt', sep='\t' )
normaldf.head()
```

```python
tumordf = tumordf.T
tumordf.rename(columns = tumordf.iloc[0], inplace= True)
tumordf = tumordf.drop(tumordf.index[0])
tumordf
```

```python
normaldf = normaldf.T
normaldf.rename(columns = normaldf.iloc[0], inplace= True)
normaldf = normaldf.drop(normaldf.index[0])
normaldf
```

```python
tumordf['label'] = 1
normaldf['label'] =0
```

```python
finaldf = pd.concat([normaldf, tumordf])

```
